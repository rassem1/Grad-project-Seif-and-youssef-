```python
# Get three integers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))

# Calculate the sum, product, and average
sum_nums = num1 + num2 + num3
product_nums = num1 * num2 * num3
average_nums = sum_nums / 3

# Print the results
print(f"\nThe sum is {sum_nums}")
print(f"\nThe product is {product_nums}")
print(f"\nThe average is {average_nums}")
```
